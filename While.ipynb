WHILE

```python
#Exemplo 1
loop = 0

while loop <= 10:
  print("Loop:", loop)
  loop += 2 # incrementar
# o loop += 2 depois do print serve para avisar que cada repetição será acrescentado mais 2
```

```python
#Exemplor 2
contador = 10

while contador >= 0:
  print("Contagem:", contador)
  contador -= 2
# -= 2 serve para dizer para decrementar menos 1
```

```python
#Exemplo 3
soma = 0
numero = 0
while numero != -1:
  numero = int(input("Digite um numero: "))
  if numero != -1:
    soma += numero
print("a soma dos numer:", soma)
```

```python
#Exemplo 4
senha_correta = "1234"
tentativa = ""

while tentativa != senha_correta:
  tentativa = input("Digite a senha: ")
print("Acesso permitido!")
```

```python
#ATIVIDADE 01
contador = 10

while contador >= 0:
  if contador == 0:
    print("Fim da contagem!")
    if contador == 0:
      break
  print("Contagem:", contador)
  contador -= 1
```

```python
#ATIVIDADE 02
numero = 0

while numero <= 20:
  if numero %2 == 0:
    print("Numero pares:", numero)
  numero += 1
```
